Data contains :- 

age - age in years
sex - (1 = male; 0 = female)
cp - chest pain type
trestbps - resting blood pressure (in mm Hg on admission to the hospital)
chol - serum cholestoral in mg/dl
fbs - (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
restecg - resting electrocardiographic results
thalach - maximum heart rate achieved
exang - exercise induced angina (1 = yes; 0 = no)
oldpeak - ST depression induced by exercise relative to rest
slope - the slope of the peak exercise ST segment
ca - number of major vessels (0-3) colored by flourosopy
thal - 3 = normal; 6 = fixed defect; 7 = reversable defect
target - have disease or not (1=yes, 0=no)

# Working of the numrical columns

```python
#we are making a divided series  of each numerical column
age_sort = df["age"].sort_values()
trestbps_sort = df["trestbps"].sort_values()
chol_sort = df["chol"].sort_values()
thalach_sort = df["thalach"].sort_values()
oldpeak_sort = df["oldpeak"].sort_values()
```

```python
#MEAN of the numerical columns
age_mean = age_sort.mean()
trestbps_mean = trestbps_sort.mean()
chol_mean = chol_sort.mean()
thalach_mean = thalach_sort.mean()
oldpeak_mean = oldpeak_sort.mean()

print(f"Mean of age: {round(age_mean,2)}")
print(f"Mean of trestbps: {round(trestbps_mean,2)}")
print(f"Mean of chol: {round(chol_mean,2)}")
print(f"Mean of thalach: {round(thalach_mean,2)}")
print(f"Mean of oldpeak: {round(oldpeak_mean,2)}")
```

```python
#VARIANCE of the numerical columns
age_var = age_sort.var()
trestbps_var = trestbps_sort.var()
chol_var = chol_sort.var()
thalach_var = thalach_sort.var()
oldpeak_var = oldpeak_sort.var()

print(f"Variance of age: {round(age_var,2)}")
print(f"Variance of trestbps: {round(trestbps_var,2)}")
print(f"Variance of chol: {round(chol_var,2)}")
print(f"Variance of thalach: {round(thalach_var,2)}")
print(f"Variance of oldpeak: {round(oldpeak_var,2)}")
```

```python
#STANDARD DEVIATION of the numerical columns
age_std = np.sqrt(age_var)
trestbps_std = np.sqrt(trestbps_var)
chol_std = np.sqrt(chol_var)
thalach_std = np.sqrt(thalach_var)
oldpeak_std = np.sqrt(oldpeak_var)

print(f"Standard Deviation of age: {round(age_std,2)}")
print(f"Standard Deviation of trestbps: {round(trestbps_std,2)}")
print(f"Standard Deviation of chol: {round(chol_std,2)}")
print(f"Standard Deviation of thalach: {round(thalach_std,2)}")
print(f"Standard Deviation of oldpeak: {round(oldpeak_std,2)}")
```

```python
age_median = np.median(age_sort)
trestbps_median = np.median(trestbps_sort)
chol_median = np.median(chol_sort)
thalach_median = np.median(thalach_sort)
oldpeak_median = np.median(oldpeak_sort)

print(f"Median of Age: {age_median}")
print(f"Median of Trestbps: {trestbps_median}")
print(f"Median of chol: {chol_median}")
print(f"Median of thalach: {thalach_median}")
print(f"Median of oldpeak: {oldpeak_median}")
```

```python
# lets find out the skewness 
age_skew = age_mean - age_median
trestbps_skew = trestbps_mean - trestbps_median
chol_skew = chol_mean - chol_median
thalach_skew = thalach_mean - thalach_median
oldpeak_skew = oldpeak_mean - oldpeak_median

print(f"Skewness of Age: {round(age_skew,2)}")
print(f"Skewness of Trestbps: {round(trestbps_skew,2)}")
print(f"Skewness of chol: {round(chol_skew,2)}")
print(f"Skewness of thalach: {round(thalach_skew,2)}")
print(f"Skewness of oldpeak: {round(oldpeak_skew,2)}")
```

# Data Preprocessing

```python
x = df.drop("target",axis=1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
print(f"X Train: {xtrain.shape}")
print(f"X Test: {xtest.shape}")
print(f"Y Train: {ytrain.shape}")
print(f"Y Test: {ytest.shape}")
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
```

```python
predict = dt.predict(xtest)
```

```python
accuracy_dt = accuracy_score(ytest,predict)
print(accuracy_dt)
```

```python
# passing our own data

my_data=np.array([[63 ,1, 3,145,233,1,0,150,0,2.3,0,0,1]])
my_data_prediction=dt.predict(my_data)
```

```python
my_data_prediction[0]
```

# KNN

```python
scaler = StandardScaler()
xtrain_std = scaler.fit_transform(xtrain)
xtest_std = scaler.transform(xtest)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k_range=range(1,26)
scores={}
scores_list=[]

for k in k_range:
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(xtrain_std,ytrain)
    prediction_knn=knn.predict(xtest_std)
    scores[k]=accuracy_score(ytest,prediction_knn)
    scores_list.append(accuracy_score(ytest,prediction_knn))
```

```python
plt.plot(k_range,scores_list)
```

```python
knn=KNeighborsClassifier(n_neighbors=6)
knn.fit(xtrain_std,ytrain)
```

```python
predict_knn=knn.predict(xtest_std)
```

```python
accuracy_knn=accuracy_score(ytest,predict_knn)
print(accuracy_knn)
```

```python
algorithms=['Decision Tree','KNN']
scores=[accuracy_dt,accuracy_knn]
```

```python
plt.figure(figsize=(5,4))
plt.xlabel("Algorithms")
plt.ylabel("Accuracy score")

sns.barplot(algorithms,scores)
```
